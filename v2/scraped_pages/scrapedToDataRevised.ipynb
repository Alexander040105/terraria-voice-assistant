```python
import time
import random
from playwright.sync_api import sync_playwright
from bs4 import BeautifulSoup
import os
import pandas as pd
```

```python
current_dir = os.getcwd()
rows = []
chunk_id = 0

for filename in os.listdir(current_dir):
    if not filename.endswith(".html"):
        continue

    page_title = filename.replace(".html", "").replace("_", " ")

    with open(filename, "r", encoding="utf-8") as f:
        soup = BeautifulSoup(f, "lxml")

    current_section = "Introduction"

    for tag in soup.find_all(["h2", "h3", "p", "li"]):
        if tag.name in ["h2", "h3"]:
            current_section = tag.get_text(strip=True)
            continue

        text = tag.get_text(strip=True)
        if not text or len(text) < 40:
            continue

        chunk_type = "paragraph" if tag.name == "p" else "list"

        rows.append({
            "id": chunk_id,
            "page_title": page_title,
            "section_title": current_section,
            "chunk_text": text,
            "chunk_type": chunk_type,
            "source_file": filename
        })

        chunk_id += 1
```

```python
df = pd.DataFrame(rows)
```

```python
df.to_csv("terraria_dataChunks.csv", index=False)
print(f"✅ Saved {len(df)} chunks to terraria_chunks.csv")
```
